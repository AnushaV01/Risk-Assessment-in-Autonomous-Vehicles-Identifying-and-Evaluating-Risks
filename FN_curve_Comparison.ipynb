```python
import matplotlib.pyplot as plt

# Data for autonomous vehicles
number_of_fatalities_av = [1, 2, 3, 4]
frequency_av = [248, 40, 7, 3]
cumulative_frequency_av = [sum(frequency_av[i:]) for i in range(len(frequency_av))]

# Data for regular vehicles
number_of_fatalities_rv = [1, 2, 3, 4, 5]
frequency_rv = [401, 38, 3, 1, 7]
cumulative_frequency_rv = [sum(frequency_rv[i:]) for i in range(len(frequency_rv))]

# Plot the FN curve
plt.figure(figsize=(10, 6))

# Autonomous vehicles FN curve
plt.plot(number_of_fatalities_av, cumulative_frequency_av, marker='o', linestyle='-', label='Autonomous Vehicles')

# Regular vehicles FN curve
plt.plot(number_of_fatalities_rv, cumulative_frequency_rv, marker='s', linestyle='--', label='Regular Vehicles')

# Labels and title
plt.xlabel('Number of Fatalities per Accident')
plt.ylabel('Cumulative Frequency of Accidents')
plt.title('FN Curve for Autonomous and Regular Vehicle Fatalities')
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for autonomous vehicles
number_of_fatalities_av = [1, 2, 3, 4]
frequency_av = [248, 40, 7, 3]
cumulative_frequency_av = [sum(frequency_av[i:]) for i in range(len(frequency_av))]

# Data for regular vehicles
number_of_fatalities_rv = [1, 2, 3, 4, 5]
frequency_rv = [401, 38, 3, 1, 7]
cumulative_frequency_rv = [sum(frequency_rv[i:]) for i in range(len(frequency_rv))]

# Plot the FN curve
plt.figure(figsize=(10, 6))

# Autonomous vehicles FN curve
plt.plot(number_of_fatalities_av, cumulative_frequency_av, marker='o', linestyle='-', label='Autonomous Vehicles')

# Regular vehicles FN curve
plt.plot(number_of_fatalities_rv, cumulative_frequency_rv, marker='s', linestyle='--', label='Regular Vehicles')

# Set log-log scale
plt.xscale('log')
plt.yscale('log')

# Labels and title
plt.xlabel('Number of Fatalities per Accident (Log Scale)')
plt.ylabel('Cumulative Frequency of Accidents (Log Scale)')
plt.title('FN Curve for Autonomous and Regular Vehicle Fatalities (Log-Log Scale)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Show plot
plt.show()
```
